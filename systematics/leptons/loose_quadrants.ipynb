```python
c = ROOT.TCanvas()
efficiencySF = ROOT.TFile('/cvmfs/atlas.cern.ch/repo/sw/database/GroupData/ElectronEfficiencyCorrection/2015_2017/rel21.2/Moriond_February2018_v2/offline/efficiencySF.offline.LooseAndBLayerLLH_d0z0_v13.root')
# efficiencySF.ls()
AtlFast2_LooseLLH_d0z0_sf_2015 = efficiencySF.Get('266904_284484/AtlFast2_LowPt_LooseLLH_d0z0_syst') 
AtlFast2_LooseLLH_d0z0_sf_2016 = efficiencySF.Get('296939_311481/AtlFast2_LowPt_LooseLLH_d0z0_syst')
AtlFast2_LooseLLH_d0z0_sf_2017 = efficiencySF.Get('324320_340453/AtlFast2_LowPt_LooseLLH_d0z0_syst')
AtlFast2_LooseLLH_d0z0_sf_2018 = efficiencySF.Get('342183_9999999/AtlFast2_LowPt_LooseLLH_d0z0_syst')
test = AtlFast2_LooseLLH_d0z0_sf_2015.Clone()
test.Add(AtlFast2_LooseLLH_d0z0_sf_2016)
test.Add(AtlFast2_LooseLLH_d0z0_sf_2017)
test.Add(AtlFast2_LooseLLH_d0z0_sf_2018)
test.Scale(1/4)
test.Draw('colz text')
# c.SetLogx()
test.SetStats(0)
test.SetTitle('AtlFast2_LowPt_LooseLLH_d0z0_syst')
test.GetXaxis().SetTitle('electron p_{T} [MeV]')
test.GetYaxis().SetTitle('electron #eta')
test.GetZaxis().SetRangeUser(0,.1)
c.Draw()
```
