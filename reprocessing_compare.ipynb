### Event number mismatch
It appears that the events used in the reprocessing are different at least for some lumiblocks.
In lumiblock 5-24 the mc event numbers do not match, but in lumiblock 25-45, there is a perfect match.
In each lumiblock, the number of events matches.

```python
for i in set(new_df.index):
    print("lumiblock", i, "has", len(new_df.loc[i]), "events")
    if not len(new_df.loc[i]) == len(old_df.loc[i]): print("event count mismatch")
    print("event numbers match:", set(new_df.loc[i].eventNumber) == set(old_df.loc[i].eventNumber))
    print()
```

### No overlap in mismatched blocks
There is no overlap in event number for the unmatched lumiblocks.
This pattern is the same for all mismatched blocks.

```python
# Mismatched block 5
print(new_df.loc[5].eventNumber.sort_values() , old_df.loc[5].eventNumber.sort_values())
```
